```python
%matplotlib inline
import sys
sys.path.append('../code')

from plotutils import *
```

```python
N = 36 # No. of trials
M = 15  # No. of gain values

alldata = []

for k1 in range(1,M+1):
    data = []
    for k2 in range(N):
        fname = '../results/Experiment5/Data/Loglikelihood-vs-Gainy/LL_Gainy' + str(k1) + '_trial' + str(k2) + '.npy'
        temp = np.load(fname)
        data.append(temp[-1])
        
    alldata.append(data)    
```

```python
reqdata = np.array(alldata)
```

```python
reqdata.shape
```

```python
for k in range(2,M+1):
    plt.plot(np.ones([N])*k,reqdata[k-1],'o',alpha=0.25,color='skyblue')

plt.plot(np.arange(2,M+1),np.mean(reqdata[1:],axis=1),'bo-')
plt.plot([0,M+1],[-370,-370],'r',linewidth=1)
plt.axis([0,M+1,-2500,0])
plt.grid(True,alpha=0.25)
plt.savefig('Loglikelihood-Gainy.pdf')
```

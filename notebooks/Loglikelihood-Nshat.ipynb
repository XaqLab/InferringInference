```python
%matplotlib inline
import sys
sys.path.append('../code')

from plotutils import *
```

```python
N = 36
M = 9

alldata10 = []
alldata5 = []
alldata1 = []

for k1 in range(2,M):
    data1 = []
    data5 = []
    data10 = []
    for k2 in range(N):
        fname1 = '../results/Experiment4/Loglikelihood-vs-Nshat-01/LL_Nshat' + str(k1) + '_trial' + str(k2) + '.npy'
        fname5 = '../results/Experiment4/Loglikelihood-vs-Nshat-05/LL_Nshat' + str(k1) + '_trial' + str(k2) + '.npy'
        fname10 = '../results/Experiment4/Loglikelihood-vs-Nshat-10/LL_Nshat' + str(k1) + '_trial' + str(k2) + '.npy'
        temp = np.load(fname1)
        data1.append(temp[-1])
        temp = np.load(fname5)
        data5.append(temp[-1])
        temp = np.load(fname10)
        data10.append(temp[-1])
    alldata1.append(data1)    
    alldata5.append(data5)    
    alldata10.append(data10)    
```

```python
reqdata1 = np.array(alldata1)
reqdata5 = np.array(alldata5)
reqdata10 = np.array(alldata10)
```

```python
fig, ax = plt.subplots(1,3,figsize=(15,4))
for k in range(2,M):
    ax[0].plot(np.ones([N])*k,reqdata1[k-2],'o',alpha=0.2,color='skyblue')
    ax[1].plot(np.ones([N])*k,reqdata5[k-2],'o',alpha=0.2,color='skyblue')
    ax[2].plot(np.ones([N])*k,reqdata10[k-2],'o',alpha=0.2,color='skyblue')
    
ax[0].plot(np.arange(2,M),np.mean(reqdata1,axis=1),'bo-')
ax[1].plot(np.arange(2,M),np.mean(reqdata5,axis=1),'bo-')
ax[2].plot(np.arange(2,M),np.mean(reqdata10,axis=1),'bo-')

ax[0].plot([1,M+1],[-705,-705],'r',linewidth=1)
ax[1].plot([1,M+1],[-370,-370],'r',linewidth=1)
ax[2].plot([1,M+1],[-317,-317],'r',linewidth=1)

ax[0].axis([1,9,-10000,0])
ax[1].axis([1,9,-3000,0])
ax[2].axis([1,9,-2000,0])

ax[0].grid(True,alpha=0.25)
ax[1].grid(True,alpha=0.25)
ax[2].grid(True,alpha=0.25)
```
